# Predicting the marks of a student based on the study hours (9.25 hours/day)

Step 1: Importing Python Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import  train_test_split
print('Libraries imported successfully!')
```

Step 2: Importing Dataset

```python
data= pd.read_csv('C:/Users/Satyaki/Documents/data.csv')
print("Importing data successfully")
```

Step 3: Check whether data imported successfully or not

Step 4: Plotting of data in graphs

```python
data.plot(x='Hours', y ='Scores', style = '1')
plt.title('Hours v/s Percentage')
plt.xlabel('Hours of Study')
plt.ylabel('Percentage of Marks')
plt.show()
```

```python
data.sort_values(["Hours"], axis = 0, ascending = [True], inplace = True)
data.plot.bar(x='Hours', y = 'Scores')
```

## After plotting the data in ascending order, we can clearly see that the percentage of marks increases with increase in study hours. This characteristic clearly resembles the real life scenario.

Now we have prepared the data for our model

```python
x= data.iloc[:, :-1].values
y= data.iloc[:, :1].values
```

Step 5: Now we have divided the data for training and testing the model

```python
from sklearn.model_selection import train_test_split
x= data['Hours']
y=data['Scores']
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.2, random_state=0)
```

## Training the Algorithm

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
model = regressor.fit(x_train.array.reshape(-1,1), y_train)
print("Training Complete")
```

Step 6: Our Model is ready. Now we have to test the model.

```python
print(x_test)
print("Prediction of Score")
y_pred=regressor.predict(x_test.array.reshape(-1,1))
y_pred
```

Step 7: Now we have to check the accuracy of our model.

```python
df = pd.DataFrame({'Actual' : y_test, 'Predicted': y_pred})
print(df) 
print("The accuracy is:" ,regressor.score(x_train.array.reshape(-1,1),y_train))
```

Step 8 :  Now we have to predic the percentage of Marks with 9.25 hours of study time per day.

```python
hours = [[9.25]]
pred= regressor.predict(hours)
print("The predicted score of the student if the study hours is 9.25 hours is :", pred)
```

## Evaluating the model

```python
from sklearn import metrics  
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred)) 
```
